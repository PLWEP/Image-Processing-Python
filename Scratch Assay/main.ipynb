```python
import matplotlib.pyplot as plt
from skimage import io
from skimage.filters.rank import entropy
from skimage.morphology import disk
import numpy as np
from scipy.stats import linregress
import glob
```

```python
img=io.imread(r"D:\Kuliah\Semester 6\Image Processing\Scratch Assay\images\Scratch0.jpg")
entropy_img = entropy(img, disk(3))
plt.imshow(entropy_img, cmap="gray")
```

```python
from skimage.filters import try_all_threshold
fig, ax = try_all_threshold(entropy_img, figsize=(10,8), verbose=False)
plt.show()
```

```python
# test isodata
from skimage.filters import threshold_isodata
def test_isodata() :
    time = 0
    time_list=[]
    area_list=[]
    path = r"D:\Kuliah\Semester 6\Image Processing\Scratch Assay\images\*.*"
    for file in glob.glob(path):
        img=io.imread(file)
        entropy_img = entropy(img, disk(3))
        thresh = threshold_isodata(entropy_img)
        binary = entropy_img <= thresh
        scratch_area = np.sum(binary == 1)
        print("time=", time, "hr  ", "Scratch area=", scratch_area, "pix\N{SUPERSCRIPT TWO}")
        time_list.append(time)
        area_list.append(scratch_area)
        time += 1

    #Print blue dots scatter plot
    plt.plot(time_list, area_list, 'bo')  

    slope, intercept, r_value, p_value, std_err = linregress(time_list, area_list)
    print("y = ",slope, "x", " + ", intercept  )
    print("R\N{SUPERSCRIPT TWO} = ", r_value**2)

test_isodata()
```
